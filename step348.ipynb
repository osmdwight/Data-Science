```python
import pandas as pd

#importing the file  using pandas

df=pd.read_csv("/content/Realestate.csv")
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

#normalizing using a built in function

df['houseAgeStandardized'] = MinMaxScaler().fit_transform(np.array(df['houseAge']).reshape(-1,1))

#dropping the column using .drop()
df.drop(columns=['numberOfConvenienceStores'], inplace=True)


df.head()
```

```python
df=df.rename(columns={'transaction':'transactionDate'}) #renaming the column
df.head()
```
