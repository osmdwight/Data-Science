```python
import seaborn as sns
df=sns.load_dataset('diamonds')
df.head()
```

```python

import matplotlib.pyplot as plt

# histogram of price
plt.figure()
sns.histplot(df['price'], kde=True)

# I chose 'price' for univariate analysis because it is a key numerical variable.
plt.show()

# The histogram shows the distribution of diamond prices and whether it is skewed
```

```python

# MULTIVARIATE ANALYSIS
# Variables chosen: 'carat' and 'price'

# plot of carat vs price
plt.figure()
sns.scatterplot(x='carat', y='price', data=df)

# I chose 'carat' and 'price' for multivariate analysis to examine their relationship.

# The scatter plot helps show how diamond weight (carat) affects price.



plt.show()
```
