```python
import matplotlib.pyplot as plt

# key and value pair
genres = {
    "Mystery": 20,
    "Western": 18,
    "Non Fiction": 35,
    "Fiction": 20,
    "Teen Fiction": 15,
    "Biography": 22,
    "Science": 15,
    "Math": 25
}

x = []
y = []
genre_names = list(genres.keys())

for i, genre in enumerate(genre_names):
    count = genres[genre]
    for j in range(count):
        x.append(i)
        y.append(j + 1)

plt.scatter(x, y)

plt.xticks(range(len(genre_names)), genre_names, rotation=45)
plt.ylabel("Number of Books")
plt.xlabel("Genre")
plt.title("Dot Plot of Books by Genre")

plt.tight_layout()
plt.show()
```
